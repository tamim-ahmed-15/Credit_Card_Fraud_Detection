# Importing Pandas and Loading dataset

```python
import pandas as pd

#dataset loading
data = pd.read_csv('creditcard.csv')
data.head(10) #shows only first 10 rows
```

# Inspecting Data

```python
print("Dataset Shape:", data.shape)
print("\nColumn Names:\n", data.columns)
print("\nData Types:\n", data.dtypes)
```

```python
print("\nSummary Statistics:\n", data.describe())
```

```python
#finding null values
print("\nMissing Values:\n", data.isnull().sum())
```

```python
print("\nClass Distribution:\n", data['Class'].value_counts())
```

# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

# 1. Histograms for Amount and Time

```python
plt.figure(figsize=(12, 5))

# Histogram for Amount
plt.subplot(1, 2, 1)
sns.histplot(data['Amount'], bins=50, kde=True, color='blue')
plt.title('Distribution of Transaction Amount')
plt.xlabel('Amount ($)')
plt.ylabel('Count')

# Histogram for Time
plt.subplot(1, 2, 2)
sns.histplot(data['Time'], bins=50, kde=True, color='green')
plt.title('Distribution of Transaction Time')
plt.xlabel('Time (seconds)')
plt.ylabel('Count')

plt.tight_layout()
plt.savefig('histograms.png')  
plt.show()
```

# 2. Correlation Matrix

```python
plt.figure(figsize=(10, 8))
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, cmap='coolwarm', center=0, annot=False)
plt.title('Correlation Matrix of Features')
plt.savefig('correlation_matrix.png')  
plt.show()
```

# Counting Normal and Fraude Transaction

```python
# Counting transactions by class
transaction_counts = data['Class'].value_counts()

# Calculating percentages
total_transactions = len(data)
transaction_percentages = (transaction_counts / total_transactions) * 100

# Normal transactions
print(f"Number of normal transactions: {transaction_counts[0]} ({transaction_percentages[0]:.2f}%)")

# Fraud Transactions
print(f"Number of fraud transactions: {transaction_counts[1]} ({transaction_percentages[1]:.2f}%)")
```

# 3. Fraud vs. Normal Distribution

```python
# Count Plot with Log Scale and Annotations
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=data, x='Class', palette='Set1')
plt.title("Fraud (1) vs Normal (0) Transaction Count (Log Scale)")
plt.xlabel("Class")
plt.ylabel("Log Count")
plt.xticks([0, 1], ['Normal', 'Fraud'])
plt.yscale('log')  

# Counting Annotate
for p in ax.patches:
    height = p.get_height()
    ax.annotate(f'{int(height)}', (p.get_x() + p.get_width() / 2., height),
                ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig('count_plot_log.png') 
plt.show()
```
